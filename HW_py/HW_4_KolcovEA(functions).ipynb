# *выполнение основного задания:*

````python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]


directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

````

````python
# 1 - Пользователь по команде "p" может узнать владельца документа по его номеру

def owner_p():
  input_number = str(input('Введите номер документа: '))
  count = 0
  for document in documents:
    if document['number'] == input_number:
      print(f'Владелец документа: {document["name"]}')
    else:
      count += 1

    if count == len(documents):
      print('Документ не найден в базе')

#owner_p()
````

````python
# 2 - Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

def shelf_s():                                      # ввод номера и перенаправление на функцию shelf_print
  input_number = str(input('Введите номер документа: '))
  shelf_print(input_number)



def shelf_return(input_n):                          # функция возвращает полку по введенному номеру или 'None'
  count = 0
  for directory in directories:
    if input_n in directories[directory]:
      return directory
    else:
      count += 1
  if count == len(directories):
    return 'None'



def shelf_print(input_n):                           # вывод результата
  if shelf_return(input_n) == 'None':
    print('Документ не найден в базе')
  else:
    print('Документ хранится на полке:', shelf_return(input_n))


#shelf_s()
````

````python
# 3 - Пользователь по команде "l" может увидеть полную информацию по всем документам

def list_l():
  for document in documents:
    print(f'№: {document["number"]}, тип: {document["type"]}, владелец: {document["name"]}, полка хранения: {shelf_return(document["number"])}')

#list_l()
````

````python
# 4 - Пользователь по команде "ads" может добавить новую полку

def add_shelf_ads():
  input_shelf = str(input('Введите номер полки: '))
  if input_shelf in directories:
    print('Такая полка уже существует. Текущий перечень полок:', ', '.join(directories.keys()))
  else:
    directories[input_shelf] = []
    print('Полка добавлена. Текущий перечень полок:', ', '.join(directories.keys()))

#add_shelf_ads()
````

````python
# 5 - Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

def delete_shelf_ds():
  input_shelf = str(input('Введите номер полки: '))
  if input_shelf in directories:
    if directories[input_shelf] == []:
      del directories[input_shelf]
      print('Полка удалена. Текущий перечень полок:', ', '.join(directories.keys()))
    else:
      print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:', ', '.join(directories.keys()))
  else:
    print('Такой полки не существует. Текущий перечень полок:', ', '.join(directories.keys()))

#delete_shelf_ds()
````

````python
# 1. Пользователь по команде "ad" может добавить новый документ в данные

def add_doc_ad():

  number_ = str(input('Введите номер документа: '))
  type_ = str(input('Введите тип документа: '))
  name_ = str(input('Введите владельца документа: '))
  shelf_ = str(input('Введите полку для хранения: '))

  if shelf_ in directories:
    new_doc = {}
    new_doc['number'] = number_
    new_doc['type'] = type_
    new_doc['name'] = name_
    documents.append(new_doc)
    directories[shelf_].append(number_)
    print('Документ добавлен. Текущий список документов: ')
  else:
    print('Такой полки не существует. Добавьте полку командой ads. Текущий список документов: ')
  list_l()

#add_doc_ad()
````

````python
# 2. Пользователь по команде "d" может удалить документ из данных

def delete_doc_d():
  number_ = str(input('Введите номер документа: '))
  count = 0
  for document in documents:
    if document['number'] == number_:
      documents.remove(document)
      print(f'Документ {number_} удален.')
    else:
      count += 1
  if count == len(documents):
    print('Документ не найден в базе.')

  print('Текущий список документов: ')
  list_l()

#delete_doc_d()
````

````python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]


directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
````

````python
# 3. Пользователь по команде "m" может переместить документ с полки на полку

def move_doc_m():
  number_ = input('Введите номер документа: ')
  shelf_ = str(input('Введите полку для перемещения: '))

  if shelf_ in directories:
    count = 0
    for document in documents:
      if document['number'] == number_:
        directories[shelf_return(number_)].remove(number_)
        directories[shelf_].append(number_)
        print(f'Документ {number_} перемещен.')
        break
      else:
        count += 1
    if count == len(documents):
      print('Документ не найден в базе.')

    print('Текущий список документов:')
    list_l()
  else:
    print('Такой полки не существует. Текущий перечень полок:', ', '.join(directories.keys()))



#move_doc_m()
````
